# 숙제 2일차

목표변수 FAT를 예측하기 위한 파생변수(BMI 등)를  생성하여 회귀분석을 실시하고 결과를 해석하라.

(BMI 계산 시, WEIGHT는 파운드(1lb = 0.45kg), HEIGHT는 인치(1inch = 2.54cm)이므로 변환하여 사용)

데이터 : 체질검사.CSV(변수: 14, 자료 수: 252개)

```python
import pandas as pd # 데이터 구성 패키지:Siries, DataFrame
import numpy as np # 행렬 연산 패키지
import matplotlib #데이터 시각화 패키지
import matplotlib.pyplot as plt
import seaborn as sns #데이터 시각화(추세 등) 패키지
from sklearn.preprocessing import scale, minmax_scale, robust_scale #Scaling클래스

from statsmodels.formula.api import ols #회귀분석
from statsmodels.api import qqplot, add_constant #qqlot, 회귀분석의 상수항 추가
import statsmodels.formula.api as smf #선형모델 formula(y~X1+X2+...)
from statsmodels.tools.eval_measures import rmse #평가함수
plt.rc('font', family=r"/home/pirl/anaconda3/lib/python3.7/site-packages/matplotlib/mpl-data/fonts/ttf/malgunbd.ttf")

from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.preprocessing import StandardScaler
#한글폰트적용
%matplotlib inline 
# matplotlib를 이용한 Background 그래프 실행 및 결과 확인
```

```python
df_raw = pd.read_csv("/home/pirl/Downloads/체질검사.csv", engine = 'python', encoding = 'EUC-KR')
df_raw.head() #데이터 불러오기
```

```python
df_raw[['WEIGHT', 'HEIGHT']] #웨이트와 헤이트 값 확인(단위 교체를 위함)
```

```python
for x in range(252): #웨이트와 헤이트의 단위 변경(각각 kg, cm)
    df_raw.loc[x:x,'WEIGHT'] = df_raw.loc[x:x,'WEIGHT'] * 0.45
    df_raw.loc[x:x,'HEIGHT'] = df_raw.loc[x:x,'HEIGHT'] * 0.0254
df_raw[['WEIGHT', 'HEIGHT']]    
    
```

```python
df_raw["BMI"] = df_raw['WEIGHT']/df_raw['HEIGHT']**2 # BMI추가 및 계산
```

```python
#이상치 확인
#웨이트와 헤이트에 비정상적인 값이 존재
df_raw.boxplot(figsize = (10,4)) 
```

```python
df_raw[df_raw["WEIGHT"]>=150 ] #150이 넘는 웨이트값 확인
```

```python
df_raw = df_raw[df_raw["WEIGHT"]<=150] #150이 넘는 웨이트 값 삭제
df_raw.boxplot(figsize = (10,4)) 
```

```python
df_raw[df_raw["HEIGHT"]<=1 ] #100이 넘지 않는 헤이트값 확인
```

```python
df_raw = df_raw[df_raw["HEIGHT"]>=1] #100이 넘지 않는 헤이트값 삭제
df_raw.boxplot(figsize = (10,4)) 
```

```python
df_raw.corr().round(3) #상관분석
```

```python
#회귀모델 생성(설명변수 전체)
reg_model = smf.ols(formula = "FAT~AGE+WEIGHT+HEIGHT+NECK+CHEST+ABDOMEN+HIP+THIGH+KNEE+ANKLE+BICEPS+FOREARM+WRIST+BMI", data=df_raw)
reg_result = reg_model.fit()
print(reg_result.summary())
#p값이 유의수준보다 작은 AGE, WEIGHT,ABDOMEN,WRIST,BMI는 유의
```

```python
#회귀모델 생성(설명변수 간의 다중공선성 확인)
df_raw_x = df_raw.drop("FAT",axis=1)
df_raw_x_const = add_constant(df_raw_x)
df_vif = pd.DataFrame()
df_vif['variable'] = df_raw_x_const.columns
df_vif["VIF"] = [variance_inflation_factor(df_raw_x_const.values,i)for i in range(df_raw_x_const.shape[1])]
df_vif.sort_values("VIF",inplace=True)
df_vif.round(3)
```

```python
#회귀모델 생성(설명변수 다중공선성 처리) 웨이트와 헤이트 제외(BMI와의 다중공선성이 높기때문)
reg_model = smf.ols(formula = "FAT~AGE+NECK+CHEST+ABDOMEN+HIP+THIGH+KNEE+ANKLE+BICEPS+FOREARM+WRIST+BMI", data=df_raw)

reg_result = reg_model.fit()
print(reg_result.summary())
```

```python
#다중공선성 높은 변수 제거 후 공선성 제확인
df_raw_x = df_raw.drop(["FAT","WEIGHT","HEIGHT"], axis = 1)
df_raw_x_const = add_constant(df_raw_x)
df_vif = pd.DataFrame()
df_vif["variable"] = df_raw_x_const.columns
df_vif["VIF"] = [variance_inflation_factor(df_raw_x_const.values,i)for i in range(df_raw_x_const.shape[1])]
df_vif.sort_values("VIF", inplace = True)
df_vif.round(3)
```

```python
#후진제거법을 이용한 변수 선택
df_raw_y = df_raw["FAT"]
model = LinearRegression()
rfe = RFE(estimator = model, n_features_to_select = 5).fit(df_raw_x, df_raw_y)
selected_cols = df_raw_x.columns[rfe.support_]
removed_cols = df_raw_x.columns[~rfe.support_]

print("Selected Variables:{}".format(selected_cols))
print("Removed Variables:{}".format(removed_cols))
```

```python
#후진제거법을 이용한 변수 선택 및 모델 생성
rfe_reg_model = smf.ols(formula = "FAT ~ NECK + ABDOMEN + HIP + WRIST + BMI ", data = df_raw)
rfe_reg_result = rfe_reg_model.fit()
print(rfe_reg_result.summary())
```

```python
#후진제거법에 의하여 선택된 데이터
df_raw_x_rfe = df_raw_x[selected_cols]
df_raw_x_rfe_const = add_constant(df_raw_x_rfe)

df_rfe_vif = pd.DataFrame()
df_rfe_vif["variable"] = df_raw_x_rfe_const.columns
df_rfe_vif["VIF"] = [variance_inflation_factor(df_raw_x_rfe_const.values, i) for i in range(df_raw_x_rfe_const.shape[1])]
df_rfe_vif.sort_values("VIF",inplace = True)
df_rfe_vif.round(3)
```

```python
#변수 중요도 확인: 비표준화 회귀계수 기준
df_reg_coef = pd.DataFrame({"Coef":rfe_reg_result.params.values[1:]}, index = selected_cols)
df_reg_coef.plot.barh(y = "Coef", legend = False)
```

```python
#표준화 회귀계수를 이용한 변수 중요도 확인
scaler = StandardScaler()
cols = df_raw_x.columns

np_scaled = scaler.fit_transform(df_raw_x) #X 스케일 변환
df_scaled = pd.DataFrame(np_scaled, columns=cols)
df_scaled["FAT"] = scaler.fit_transform(df_raw[["FAT"]]) #Y스케일 변환
reg_model_scaled = smf.ols(formula = "FAT~NECK+ABDOMEN+HIP+WRIST+BMI",data=df_scaled)
reg_result_scaled = reg_model_scaled.fit()
print(reg_result_scaled.summary())
```

```python
df_reg_coef = pd.DataFrame({"Coef":reg_result_scaled.params.values[1:]}, index = selected_cols)
df_reg_coef.plot.barh(y = "Coef", legend = False)
```

회귀계수의 방향은 동일하나 중요도 크기가 변함
복부 둘레가 가장 중요하고 손목 두께의 중요도는 줄어든 반면 엉덩이 변수의 중요도도 증가함.
BMI의 중요도도 상당히 높음.
단 모델의 통계량은 변화 없음.
